```python
from transformers import pipeline
sentiment_analysis = pipeline("sentiment-analysis")

def sentiment_analyze(text):
    result = sentiment_analysis(text)
    return result

text = "I am sad to be here."
sentiment = sentiment_analyze(text)
print(sentiment)
```

```python
def get_chatbot_response(user_input):
    """Simple rule-based chatbot that responds to user input"""
    user_input = user_input.lower().strip()
    
    # Greeting rules
    if user_input in ['hello', 'hi', 'hey', 'greetings']:
        return "Hello! How can I help you today?"
    
    # How are you rules
    elif user_input in ['how are you', 'how are you?', "how's it going"]:
        return "I'm doing great, thanks for asking! How about you?"
    
    # Help rules
    elif 'help' in user_input or user_input == '?':
        return "I can help with greetings, ask about the weather, tell jokes, or answer simple questions!"
    
    # Weather rules
    elif 'weather' in user_input or 'temperature' in user_input:
        return "I don't have real-time weather data, but you can check a weather website!"
    
    # Joke rules
    elif 'joke' in user_input or 'funny' in user_input:
        return "Why did the programmer quit his job? Because he didn't get arrays!"
    
    # Goodbye rules
    elif user_input in ['bye', 'goodbye', 'see you', 'farewell']:
        return "Goodbye! Have a great day!"
    
    # Default response
    else:
        return "I'm not sure how to respond to that. Try saying 'hello' or 'help'!"


def main():
    """Main chatbot loop"""
    print("=== Simple Rule-Based Chatbot ===")
    print("Type 'quit' to exit\n")
    
    while True:
        user_input = input("You: ").strip()
        
        if user_input.lower() == 'quit':
            print("Chatbot: Goodbye!")
            break
        
        if not user_input:
            continue
        
        response = get_chatbot_response(user_input)
        print(f"Chatbot: {response}\n")


if __name__ == "__main__":
    main()
```
