```python
import nltk
from nltk import CFG
import spacy
```

```python
grammar = CFG.fromstring("""
S -> NP VP
NP -> DT NN
VP -> VBZ NP | VBZ JJ
DT -> 'The' | 'A'
NN -> 'cat' | 'dog'
VBZ -> 'is' | 'likes'
JJ -> 'happy' | 'big'
""")
```

```python
parser = nltk.ChartParser(grammar)
```

```python
def check_grammar(sentence):
    tokens = sentence.split()
    try:
        trees = list(parser.parse(tokens))
        if trees:
            print(f"'{sentence}' is grammatically correct.")
            for tree in trees:
                tree.pretty_print()    # NLTK's tree visual
        else:
            print(f"'{sentence}' is NOT grammatically correct.")
    except Exception as e:
        print(f"Error parsing sentence: {e}")

check_grammar("The dog is happy")
```

```python
check_grammar("A cat likes dog")
```

```python
check_grammar("The dog run fast")  # fails
```

```python
nlp = spacy.load("en_core_web_sm")

# Input text
text = "The quick brown fox jumps over the lazy dog."

# Process the text
doc = nlp(text)

# Part-of-Speech (POS) Tagging
print("Part-of-Speech (POS) Tagging:")
for token in doc:
    print(f"Word: {token.text:<10} POS: {token.pos_:<5} Tag: {token.tag_:<5} Dependency: {token.dep_}")

# Dependency Parsing
print("\nSyntactic Parsing (Dependency Analysis):")
for token in doc:
    children = [child.text for child in token.children]
    print(f"Word: {token.text:<10} Dependency: {token.dep_:<8} Head: {token.head.text:<10} Children: {children}")
```
