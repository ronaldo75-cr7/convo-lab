```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics import classification_report
```

```python
text = ["Thomson is gay.",
       "Whatever the sun is.",
       "Bro knows something that we dont.",
       "Samuel jose kutty is a bodybuilder.",
       "9/11 was not an accident.",
       "Obama Bih laden was arrested for nothing.",
       "There is always an asian kid that is better than u."]
labels = [0,1,0,0,1,0,1]

X_train, X_test, y_train, y_test = train_test_split(text, labels, test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
clf = MultinomialNB()
clf.fit(X_train_tfidf, y_train)

# Make predictions and evaluate the classifier
y_pred = clf.predict(X_test_tfidf)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer


docs = [
    "Natural language processing enables machines to understand human language.",
    "Text classification is a fundamental task in NLP.",
    "Clustering of text data is useful for organizing documents."
]


count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(docs)
print("Count Vectorizer Features:\n", X_count.toarray())


tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(docs)
print("TF-IDF Features:\n", X_tfidf.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
docs = [
    "Thomson is gay.",
       "Whatever the sun is.",
       "Bro knows something that we dont.",
       "Samuel jose kutty is a bodybuilder.",
       "9/11 was not an accident.",
       "Obama Bih laden was arrested for nothing.",
       "There is always an asian kid that is better than u."
]
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(docs)
num_clusters = 2  
kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
kmeans.fit(X_tfidf)
print("Cluster Labels for Documents:")
for i, label in enumerate(kmeans.labels_):
    print(f"Document {i} → Cluster {label}: {docs[i]}")
terms = tfidf_vectorizer.get_feature_names_out()
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]
print("\nTop words per cluster:")
for i in range(num_clusters):
    print(f"Cluster {i}: ", end='')
    for ind in order_centroids[i, :5]:  
        print(terms[ind], end=' ')
    print()
```
